```python
grid = gridspec.GridSpec(3,5) 
plt.figure(figsize=(16,10))
plt.subplots_adjust(wspace= 0.3, hspace= 0.3)

features = list(df.columns[1:14])

for idx, feature in enumerate(features):
    ax = plt.subplot(grid[idx])
    
    # sns.distplot을 사용해 feature별 class 분포 확인
    sns.distplot(df[feature][df['class'] == 0], label = 0)
    sns.distplot(df[feature][df['class'] == 1], label = 1) 
    sns.distplot(df[feature][df['class'] == 2], label = 2) 
    ax.legend()
```

- color intensity는 class 1, 0, 2 순서대로 높지만, 겹치는 부분이 상대적으로 많음
- proline은 class1과 0, class2와 0은 분류가 잘 되는 편이지만, class 1, 2 사이에 분류가 잘 이루어지지 않음

정규성 검정

QQ plot

- 정규분포를 나타내는 직선을 그리고, 이를 실제 데이터와 비교
- 직선과 데이터의 분포가 일치할수록 정규분포를 따름
- 정확한 p-value 값을 알 수 없다는 한계 존재

```python
from scipy import stats
```

```python
grid = gridspec.GridSpec(3,5) 
plt.figure(figsize=(16,10))
plt.subplots_adjust(wspace= 0.3, hspace= 0.3)

for idx, feature in enumerate(features) : 
    ax = plt.subplot(grid[idx])
    stats.probplot(df[feature], plot = plt)
    ax.set_title(feature)
```

- alcohol, ash, alcalinity_of_ash, magnesium, total_phenols, nonflavanoid_phenols, proanthocyanins, hue가 정규분포를 따름
- feature 분포 그래프와 비교했을 때 거의 유사함

shapiro test
- p-value : 귀무가설이 맞다고 가정할 때 얻은 결과보다 극단적인 결과가 관측될 확률. 일반적으로 p-value가 0.05 미만일 때 귀무가설을 기각하고 0.05 이상일 때 귀무가설을 채택
- 귀무가설을 '정규분포를 따른다'로, 대립가설을 '정규분포를 따르지 않는다'고 가정함
- 즉, p-value가 0.05 이상인 경우 정규분포를 따름

```python
for i in features :
    test_stat, p_val = stats.shapiro(df[i])
    print(f' {i} : {p_val}')
```

- alcohol, non-flavanoids만 정규분포를 따름
- QQ plot, feature 분포 그래프를 통해 예측한 것과 비교해보았을 때 더 엄격히 정규성 검정을 실시함을 알 수 있음
