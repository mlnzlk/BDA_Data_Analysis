### Concat 함수
- df + df 끼리 합치는 경우 ( 두 개의 테이블이 행과 열 기준으로 붙여진다. )
- axis 축,  ignore_index 병합후 인덱스 새롭게 만들지 유무

```python
import pandas as pd

df1 = pd.DataFrame({'A':[1,2,3,4],'B':[1,2,3,4]})
df2 = pd.DataFrame({'A':[4,5],'B':[4,5]})
df3 = pd.DataFrame({'A':[5,6],'C':[3,2]})

display(df1)
display(df2)
display(df3)
```

```python
pd.concat([df1,df2,df3],ignore_index=True) #axis= 0 디폴트값
```

```python
pd.concat([df1,df3,df2],ignore_index=True) #axis= 0 디폴트값
```

```python
pd.concat([df1,df3,df2],axis=1) #axis= 0 디폴트값
```

- concat의 경우는 컬럼이 동일한 경우는 쉽게 붙일 수 있다.
- 다만 시계열데이터는 다를 수 있으니 유의하기!

## merge 함수
- 두 개의 데이터프레임을 join하는 함수 
- 공통된 컬럼을 기준으로 데이터프레임을 병합하는 것이 가능하다.
- 공통된 컬럼이 join key
- 함수 주요 인자
- left : 병합할 첫 번째 데이터 프레임으로 왼쪽 ( 순서가 기준 )
- right : 병합할 두 번째 데이터 프레임으로, 오른쪽 데이터 프레임 
- on : 조인 키로 사용할 컬럼 목록으로, 두 데이터프레임 모두 포함돼어야 한다.
- left_on : 왼쪽 데이터프레임에서 조인 키로 사용할 컬럼 목록
- right_on : 오른쪽 데이터프레임에서 조인키로 사용할 컬럼 목록
- left_index : 왼쪽 데이터프레임의 조인 키로 인덱스를 사용하지
- right_idnex: 오른쪽 데이터프레임의 조인키로 인덱스를 사용할지
- how : 조인의 방법을 결정하며, 디폴트는 'inner'

```python
df1 = pd.DataFrame({'이름':['홍길동','김영희','김철수','임꺽정'],
                   '영어':[100,50,60,70]})
df2 = pd.DataFrame({'이름':['홍길동','김영희','김철수'],
                   '수학':[50,100,60]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2) #디폴트는 inner 교집합의 개념으로 들어감 
```

```python
pd.merge(df1,df2) #디폴트는 inner 교집합의 개념으로 들어감 
```

```python
pd.merge(df1,df2, on='이름')
```

```python
df3 = pd.DataFrame({'이름':['홍길동','김영희','김철수','임꺽정'],
                   '수학':[50,100,60,70],
                   '생년월일':['00','01','02','03']})
df4 = pd.DataFrame({'이름':['홍길동','김영희','김철수'],
                   '수학':[50,100,60],
                   '지각횟수':[0,1,2]})
```

- left_on, right_on 

```python
df3 = pd.DataFrame({'이름':['홍길동','김영희','김철수','임꺽정'],
                   '수학':[50,100,60,70],
                   '생년월일':['00','01','02','03']})
df4 = pd.DataFrame({'name':['홍길동','김영희','김철수'],
                   '수학':[50,100,60],
                   '지각횟수':[0,1,2]})
```

```python
pd.merge(df3,df4, left_on=['이름'],right_on=['name'])
```

```python
pd.merge(df3,df4, left_on=['이름','수학'],right_on=['name','수학'])
```

- left_index, right_index

```python
df3_sp.index
```

```python
df3_sp
```

```python
df4
```

```python
pd.merge(df3_sp, df4, left_index=True, right_on='name')
```

### How
- 조인 키에 대한 inner 기본값
- left :왼쪽 데이터프레임 조인 키를 기준 병합
- right : 오른쪽 데이터프레임의 조인 키를 기준 병합
- outter : 왼쪽과 오른쪽 데이터프레임 조인키에 대한 합집합 기준 

```python
df1 = pd.DataFrame({'A':[1,2,3], 'B':[1,2,3]})
df2 = pd.DataFrame({'A':[2,3,4], 'C':[7,8,9]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on='A',how = 'inner')
```

```python
pd.merge(df1, df2, on='A')
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on='A',how = 'left')
```

```python
pd.merge(df1, df2, on='A',how = 'right')
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on='A',how = 'outer')
```
