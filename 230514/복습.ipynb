데이터 병합

concat
- 데이터프레임 병합 (df+df)
- axis 축, ignore_index 병합 후 새로운 인덱스 생성 여부

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A':[1,2,3,4],'B':[1,2,3,4]})
df2 = pd.DataFrame({'A':[4,5],'B':[4,5]})
df3 = pd.DataFrame({'A':[5,6],'C':[3,2]})
```

```python
display(df1)
display(df2)
display(df3)
```

```python
pd.concat([df1,df2,df3],ignore_index = True) 
# axis = 0 열 기준 병합
```

```python
pd.concat([df1,df3,df2],ignore_index = True) 
# 입력한 순서대로 병합 (결측치 위치 다름)
```

```python
pd.concat([df1,df2,df3], axis = 1) # 행 기준 병합
```

merge
- 두 개의 데이터프레임을 공통된 컬럼을 기준으로 병합
- join key
- left : 첫 번째(왼쪽) 데이터 프레임
- right : 두 번째(오른쪽) 데이터 프레임
- on : 조인 키로 사용할 컬럼 목록으로, 두 데이터 프레임에 공통된 데이터만을 병합
- left_on : 첫 번째 데이터 프레임에서 조인 키로 사용할 컬럼 목록
- right_on : 두 번째 데이터 프레임에서 조인 키로 사용할 컬럼 목록
- left_index : 첫 번째 데이터프레임의 인덱스를 조인키로 사용
- right_index : 두 번째 데이터프레임의 인덱스를 조인키로 사용
- how : 조인 방법 결정 (inner, outer, left, right / 디폴트는 inner)

```python
df1 = pd.DataFrame({'Name' : ['Kim', 'Kang', 'Lee', 'Park'],
                   'English' : [80, 90, 75, 60]})
df2 = pd.DataFrame({'Name' : ['Kim', 'Kang', 'Lee'],
                   'Math' : [70, 85, 90]})
```

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2) 
# 두 데이터 프레임에 모두 포함된 데이터에 대해 병합 (inner)
```

```python
pd.merge(df1, df2, on = 'Name')
# Name열을 기준으로 공통된 데이터(Kim, Kang, Lee)에 대해 병합
```

```python
df3 = pd.DataFrame({'Name' : ['Kim', 'Kang', 'Lee', 'Park'],
                   'English' : [80, 90, 75, 60],
                   'Birth' : ['00', '01', '02', '03']})
df4 = pd.DataFrame({'Name' : ['Kim', 'Kang', 'Lee'],
                   'English' : [80, 90, 75],
                   'Class' : ['A', 'B', 'C']})
```

```python
display(df3)
display(df4)
```

```python
pd.merge(df3, df4)
```

- on

```python
pd.merge(df3, df4, on = ['Name', 'English']) # 두 개 이상 컬럼 사용 가능
```

- left_on, right_on

```python
df3 = pd.DataFrame({'Name' : ['Kim', 'Kang', 'Lee', 'Park'],
                   'English' : [80, 90, 75, 60],
                   'Birth' : ['00', '01', '02', '03']})
df4 = pd.DataFrame({'Names' : ['Kim', 'Kang', 'Lee'],
                   'English' : [80, 90, 75],
                   'Class' : ['A', 'B', 'C']})
```

```python
pd.merge(df3, df4, left_on = ['Name'], right_on = ['Names'])
# df3의 Name, df4의 Names를 기준으로 공통된 데이터만 가져와 병합
```

```python
pd.merge(df3, df4, left_on = ['Name', 'English'], right_on = ['Names', 'English'])
```

- left_index, right_index

```python
df3_sp = df3.set_index('Name')
# df3의 Name 컬럼을 인덱스로 지정
```

```python
df3_sp.index
```

```python
df3_sp
```

```python
df4
```

```python
pd.merge(df3_sp, df4, left_index = True, right_on = 'Names')
# df3_sp의 인덱스와 df4의 Names 컬럼을 기준으로 공통된 데이터만 병합
```

- how
- left : 왼쪽 데이터프레임 컬럼을 기준으로 병합
- right : 오른쪽 데이터프레임 컬럼을 기준으로 병합
- inner : 두 데이터 프레임에 공통된 컬럼 사용
- outer : 두 데이터 프레임 모든 컬럼 사용

```python
df1 = pd.DataFrame({'A':[1,2,3], 'B':[1,2,3]})
df2 = pd.DataFrame({'A':[2,3,4], 'C':[7,8,9]})
```

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, on = 'A', how = 'inner')
# A컬럼을 기준으로 공통된 데이터만 병합
# inner가 디폴트
```

```python
pd.merge(df1, df2, on = 'A', how = 'left')
# 왼쪽 데이터프레임(df1)의 A 컬럼 기준으로 병합
```

```python
pd.merge(df1, df2, on = 'A', how = 'right')
# 오른쪽 데이터프레임(df12)의 A 컬럼 기준으로 병합
```

```python
pd.merge(df1, df2, on = 'A', how = 'outer')
# A컬럼을 기준으로 모든 데이터 병합
```
