```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style = 'whitegrid')

# 그래프 크기 설정
plt.figure(figsize = (5,5))

# 산점도 그래프 생성

a = np.random.normal(0,1,100)
b = np.random.normal(0,1,100)
x = np.random.normal(0,1,100)
y = np.random.normal(0,1,100)

plt.scatter(a, b, c = 'blue', marker = 'o', alpha = 0.4, linewidths = 1, label = 'Group A')
plt.scatter(x, y, c = 'green', marker = 'D', alpha = 0.4, linewidths = 1, label = 'Group B')

# 그래프 이름, 축 설정
plt.title('score')
plt.xlabel('English')
plt.ylabel('Math')

# 축 범위 설정
plt.xlim(-5,5)
plt.ylim(-5,5)

# 눈금값 조절, 범례 설정
plt.xticks([-2,0,2], ['C', 'B', 'A'])
plt.yticks([-2,0,2], ['C', 'B', 'A'])
plt.legend()
```
