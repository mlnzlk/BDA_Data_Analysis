```python
import matplotlib.pyplot as plt
import seaborn as sns
```

마커 설정

- . (point)

```python
sns.set(style = 'white')
plt.plot([1,2,3], [4,5,6], marker = '.', label = 'cty', color = 'r', linestyle = 'dashed', linewidth = 1)
plt.legend()
```

- o (circle)

```python
sns.set(style = 'whitegrid')
plt.plot([4,5,6], [7,8,9], marker = 'o', label = 'hwy', color = 'g', linestyle = 'dotted', linewidth = 2)
plt.legend()
```

- D (Diamond)

```python
sns.set(style = 'dark')
plt.plot([2,4,3], [4,5,3], marker = 'D', label = 'displ', color = 'b', linestyle = 'dashdot', linewidth = 2)
plt.legend()
```

- s (square)

```python
sns.set(style = 'darkgrid')
plt.plot([7,8,9], [4,7,11], marker = 's', label = 'cyl', color = 'black', linestyle = 'solid', linewidth = 2)
plt.legend()
```

- p (plus)

```python
sns.set(style = 'ticks')
plt.plot([2,4,6], [5,3,7], marker = 'p', label = 'score', color = 'yellow', linestyle = 'dotted', linewidth = 2)
plt.legend()
```

- ^ (triangle up)

```python
sns.set(style = 'whitegrid')
plt.plot([3,5,2], [4,6,7], marker = '^', label = 'count', color = 'gray', linestyle = 'dashed', linewidth = 2)
plt.legend()
```
