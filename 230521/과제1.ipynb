```python
crime = pd.read_csv('crime.csv')
```

#### 1. 시계열 데이터로 데이터 형태 변경하기 **to_datatime()**

```python
crime.info()
```

```python
crime['REPORTED_DATE'] = pd.to_datetime(crime['REPORTED_DATE'])
```

```python
crime.info()
```

- REPORTED_DATE 열의 데이터 형태가 object에서 datetime64로 변경된 것을 확인 가능
- 이렇게 데이터 형태를 시계열로 변경한 후 분석을 진행해야 시계열 데이터 분석 자체가 가능함

- 시계열 데이터 컬럼을 인덱스로 지정

```python
crime_dt = crime.set_index('REPORTED_DATE')
```

### 2. 처음/끝 특정기간 필터링 **first/last**
- 처음/끝 값으로부터 특정 기간만큼의 데이터를 필터링하는 방법
- df.first(기준이 되는 기간)
- df.last(기준이 되는 기간)

- 시계열 데이터 시간별로 정렬

```python
crime_dt_sort = crime_dt.sort_index()
```

- 첫 4일 데이터 추출

```python
crime_dt_sort.first('4D')
```

- 마지막 4일 데이터 추출

```python
crime_dt_sort.last('4D')
```

- 정렬 후 추출해야 값이 나옴 (정렬하지 않으면 error 발생)

### 3. 특정시간 필터링 **at_time**
- df.at_time(time, axis = None)
- 특정 시간만 추출
- time : 기준이 될 시간
- axis : 기준이 될 축

- 2017년의 11시 54분 데이터만 추출

```python
crime_dt17 = crime_dt.loc['2017'].sort_index()
```

```python
crime_dt17.at_time('11:54')
```

### 4. 특정기간 필터링 **between_time**
- df.between_time(start_time, end_time, include_start = True, include_end = True, axis = None)
- start_time : 기준이 될 시작 시간
- end_time : 기준이 될 끝 시간
- include_start : 시작 시간을 포함할지 여부
- include_end : 끝 시간을 포함할지 여부
axis : 기준이 될 축

- 2015년의 1시~4시 데이터만 추출

```python
crime_dt15 = crime_dt.loc['2015'].sort_index()
```

```python
# 1. 시작, 끝 모두 포함
crime_dt17.between_time('1:00', '4:00')
```

```python
# 2. 시작 포함, 끝 제외
crime_dt17.between_time('1:00', '4:00', include_end = False)
```

```python
# 3. 시작 제외, 끝 포함
crime_dt17.between_time('1:00', '4:00', include_start = False)
```

```python
# 4. 시작, 끝 모두 제외
crime_dt17.between_time('1:00', '4:00', include_start = False, include_end = False)
```

- 조건 설정에 따라 행 수가 달라지는 것 확인가능

### 5. 데이터나 인덱스를 원하는 기간만큼 쉬프트해주기 **shift**
- df.shift(periods = 1, freq = None, axis = 0, fill_value = NoDefault.no_default)
- periods : 이동할 기간
- freq : 입력 할 경우 인덱스가 이동  
Y, M, D, H, T, S 나 Timestamp, 'Infer'등 가능
- fill_value : shift로 인해 생긴 결측치 대체하는 값

```python
crime_dt_sort.shift(periods = 5, freq = 'Y')
```

- 년 기준 5기간 이동. 즉, 5일 이동

### 6. period로 변환 **to_period**
- df.to_period(freq = None, axis = 0, copy = True)
- freq : 원하는 시간 단위로 변환
- axis : 변환 할 기준 축
- copy : 사본을 형성할지 여부

- 원하는 시간 간격으로 출력

```python
# 년 기준
crime_dt_sort.to_period("Y")
```

```python
# 월 기준
crime_dt_sort.to_period("M")
```

```python
# 주 기준
crime_dt_sort.to_period("W")
```
